```python
# dataframe of just tuna

dftuna = df.drop('img_id', axis=1)
dftuna = dftuna.drop('label_l2', axis=1)
dftuna = dftuna[dftuna['label_l1'].isin(['Albacore', 'Bigeye tuna', 'Skipjack tuna', 'Yellowfin tuna'])]
print(dftuna.head())
```

```python
# get summary statistics of just tuna

print(dftuna.describe())
```

```python
#Bar plot of tuna count

ax = sns.countplot(dftuna, x="label_l1", color="green")
ax.set_title('Distribution of Tuna Types', fontsize=15)
ax.set_xlabel('Tuna Species', fontsize=12)
ax.set_ylabel('Count', fontsize=12)

#with count labels
for counts in ax.patches:
    ax.annotate(f'{counts.get_height()}',
                (counts.get_x() + counts.get_width()/2., counts.get_height()),
                ha='center', va='bottom')
```

```python
# Pie chart for proportion

tuna_counts = dftuna['label_l1'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(tuna_counts, labels=tuna_counts.index, autopct='%1.1f%%', 
        startangle=90, shadow=True, explode=[0.01]*len(tuna_counts))
plt.title('Proportion of Tuna Types', fontsize=15)
plt.axis('equal')
plt.tight_layout()
plt.show()
```

```python
#Count all species types
all_count = df['label_l1'].value_counts()

#Total tuna count

tuna_count = df[df['label_l1'].isin(['Albacore', 'Bigeye tuna', 'Skipjack tuna', 'Yellowfin tuna'])].shape[0]
other_count = df.shape[0] - tuna_count

#summary df for distrib

dfsumm = pd.DataFrame({
    'Category': ['Tuna', 'Other Species'],
    'Count': [tuna_count, other_count]
})

#get Count by Category
print(dfsumm)
```

```python
#Bar plot showing tuna vs other species

plt.figure(figsize=(10, 6))
ax = sns.barplot(x='Category', y='Count', data=dfsumm)
ax.set_title('Tuna vs Other Species Count', fontsize=15)
for count in ax.patches:
        ax.annotate(f'{count.get_height()}', 
                (count.get_x() + count.get_width()/2., count.get_height()),
                ha='center', va='bottom')
plt.tight_layout()
plt.show()
```

```python
# Detailed distrib of top 10 species

plt.figure(figsize=(14, 8))
top_species = all_count.head(10)

#pallette for differentiation
ax = sns.barplot(x=top_species.index, y=top_species.values, palette='viridis')
ax.set_title('Top 10 Species by Count', fontsize=15)
ax.set_xlabel('Species', fontsize=12)
ax.set_ylabel('Count', fontsize=12)
plt.xticks(rotation=45, ha='right')
for count in ax.patches:
    ax.annotate(f'{count.get_height()}', 
                (count.get_x() + count.get_width()/2., count.get_height()),
                ha='center', va='bottom')
plt.tight_layout()
plt.show()
```

```python
# Create a function to calculate bounding box center
def add_bbox_centers(df):
    df = df.copy()
    df['center_x'] = (df['x_min'] + df['x_max']) / 2
    df['center_y'] = (df['y_min'] + df['y_max']) / 2
    df['width'] = df['x_max'] - df['x_min']
    df['height'] = df['y_max'] - df['y_min']
    df['area'] = df['width'] * df['height']
    return df
```

```python
# Apply to both dfs
df_with_centers = add_bbox_centers(df)
dftuna_with_centers = add_bbox_centers(dftuna)
tuna_species = ['Albacore', 'Bigeye tuna', 'Skipjack tuna', 'Yellowfin tuna']
```

```python
# Scatter plot of bounding box centers for ALL species
plt.figure(figsize=(10, 8))
plt.scatter(df_with_centers['center_x'], df_with_centers['center_y'], 
           alpha=0.1, s=5, c='blue')
plt.title('Distribution of All Bounding Box Centers', fontsize=15)
plt.xlabel('X coordinate', fontsize=12)
plt.ylabel('Y coordinate', fontsize=12)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
#Scatter plot of bounding box centers for tuna species
plt.figure(figsize=(12, 10))
for species, color in zip(tuna_species, ['red', 'green', 'blue', 'purple']):
    species_data = dftuna_with_centers[dftuna_with_centers['label_l1'] == species]
    plt.scatter(species_data['center_x'], species_data['center_y'], 
                alpha=0.3, s=10, label=species)
plt.title('Distribution of Tuna Bounding Box Centers by Species', fontsize=15)
plt.xlabel('X coordinate', fontsize=12)
plt.ylabel('Y coordinate', fontsize=12)
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# Heatmap of bounding box distribs for each tuna type
plt.figure(figsize=(20, 5))
for i, species in enumerate(tuna_species, 1):
    plt.subplot(1, 4, i)
    species_data = dftuna_with_centers[dftuna_with_centers['label_l1'] == species]
    sns.kdeplot(x='center_x', y='center_y', data=species_data, 
                fill=True, cmap='viridis', alpha=0.7)
    plt.title(f'{species} Locations', fontsize=12)
    plt.xlabel('X coordinate')
    plt.ylabel('Y coordinate')
plt.tight_layout()
plt.show()
```

```python
#box plots for bounding box sizes by species

#width
plt.figure(figsize=(14, 7))
plt.subplot(1, 2, 1)
sns.boxplot(x='label_l1', y='width', data=dftuna_with_centers)
plt.title('Bounding Box Width by Tuna Species', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.ylabel('Width (pixels)')

# height
plt.subplot(1, 2, 2)
sns.boxplot(x='label_l1', y='height', data=dftuna_with_centers)
plt.title('Bounding Box Height by Tuna Species', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.ylabel('Height (pixels)')
plt.tight_layout()
plt.show()
```

```python
# Area comparisons
plt.figure(figsize=(10, 6))
sns.boxplot(x='label_l1', y='area', data=dftuna_with_centers)
plt.title('Bounding Box Area by Tuna Species', fontsize=15)
plt.xticks(rotation=45, ha='right')
plt.ylabel('Area (pixels²)')
plt.tight_layout()
plt.show()
```
