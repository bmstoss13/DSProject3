```python
import pandas as pd 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("cleaned_foid (1).csv")
df.head()
```

```python
# missing values

print(df.isna().sum())
len(df)
```

```python
# label
frequency_table = df['label'].value_counts().reset_index()
frequency_table.columns = ['Value', 'Frequency']
print(frequency_table)


counts = df['label'].value_counts()
counts.plot(kind='bar', color='skyblue')
plt.xlabel('Label')
plt.ylabel('Count')
plt.title('Value Counts of Labels')
plt.xticks(rotation=45, ha='right') 
plt.show()
```

```python
# center_x

print(df['center_x'].describe())


plt.hist(df['center_x'], bins=30, color='skyblue', edgecolor='black')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of center_x')
plt.show()
```

```python
# center_y

print(df['center_y'].describe())


plt.hist(df['center_y'], bins=30, color='skyblue', edgecolor='black')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of center_y')
plt.show()
```

```python
# width

print(df['width'].describe())


plt.hist(df['width'], bins=30, color='skyblue', edgecolor='black')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of Bounding Box Width')
plt.show()
```

```python
# height

print(df['height'].describe())

plt.hist(df['height'], bins=30, color='skyblue', edgecolor='black')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of Bounding Box Height')
plt.show()
```
